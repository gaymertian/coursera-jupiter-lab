```python
import folium
from folium.features import DivIcon

# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initialize the map centered at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(
    nasa_coordinate,
    radius=1000,  # Radius in meters
    color='#d35400',
    fill=True,
    fill_color='#d35400',  # Fill color
    fill_opacity=0.5
).add_child(folium.Popup('NASA Johnson Space Center'))

# Create a blue circle at NASA Johnson Space Center's coordinate with a text label
marker = folium.Marker(
    location=nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % 'NASA JSC'
    )
)

# Add the circle and marker for NASA Johnson Space Center to the map
site_map.add_child(circle)
site_map.add_child(marker)

# Add a Circle and Marker for each launch site in the dataframe
for _, row in launch_sites_df.iterrows():
    site_name = row['Launch Site']
    coord = [row['Lat'], row['Long']]
    
    # Add Circle
    folium.Circle(
        location=coord,
        radius=1000,  # Radius in meters
        color='#000000',  # Border color
        fill=True,
        fill_color='#3186cc',  # Fill color
        fill_opacity=0.5
    ).add_child(folium.Popup(site_name)).add_to(site_map)
    
    # Add Marker
    folium.Marker(
        location=coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{}</b></div>'.format(site_name)
        )
    ).add_child(folium.Popup(site_name)).add_to(site_map)

# Save the map to an HTML file
site_map.save('launch_sites_map.html')  # Save the map to an HTML file

# If running in Jupyter Notebook, display the map
site_map

```

```python
import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon
import pandas as pd
from js import fetch
import io

# Fetch and read the dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Add a new column for marker colors based on the 'class' column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initialize the map centered at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Add a Circle and Marker for each launch record
for _, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    marker_color = row['marker_color']
    
    # Create and add the marker
    folium.Marker(
        location=coord,
        icon=folium.Icon(color=marker_color),
        popup=f"Launch Site: {row['Launch Site']}<br>Status: {'Success' if row['class'] == 1 else 'Failure'}"
    ).add_to(marker_cluster)

# Save the map to an HTML file
site_map.save('launch_records_map.html')  # Save the map to an HTML file

# If running in Jupyter Notebook, display the map
site_map
```

```python
import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon
import pandas as pd
from js import fetch
import io

# Fetch and read the dataset
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Add a new column for marker colors based on the 'class' column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]

# Initialize the map centered at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Add a Circle and Marker for each launch record
for _, row in spacex_df.iterrows():
    coord = [row['Lat'], row['Long']]
    marker_color = row['marker_color']
    
    # Create and add the marker
    marker = folium.Marker(
        location=coord,
        icon=folium.Icon(color=marker_color),
        popup=f"Launch Site: {row['Launch Site']}<br>Status: {'Success' if row['class'] == 1 else 'Failure'}"
    )
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
